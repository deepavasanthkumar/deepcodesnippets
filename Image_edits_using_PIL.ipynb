## Upload your image


```python
from google.colab import files
uploaded = files.upload()
filename = next(iter(uploaded))
print(f"Uploaded filename: {filename}")
```

### Use PIL to import your file and check the current file resolution

```python
from PIL import Image

# Open an image
try:
    img = Image.open(filename)
    width, height = img.size
    print(f'{filename} Image resolution is {width}/{height}')
except FileNotFoundError:
    print("Error: input.jpg not found. Please ensure the file exists.")
    exit()
```

Resize Images

```python
new_image = img.resize((400, 600))

# Save the resized image
new_image.save("resized_image.jpg")
resize_img = Image.open("resized_image.jpg")
print(f'Image resolution is {resize_img.size}')

```

### View the resized image

```python
resize_img
```

# For example, to crop a region from (250, 250) to (600, 700)

```python
crop_rectangle = (250, 250, 600, 700)
cropped_im = img.crop(crop_rectangle)
cropped_im
cropped_im.save("cropped_image.jpg")
```
